```python
df["artist"].value_counts().index.tolist()[:10]  #grabbing indices of top 10 occuring artists
```

```python
names = df.loc[df['artist'].isin(df["artist"].value_counts().index.tolist()[:10])] #making a df of these names using 
#df.loc and passing in the argument passed in above line
```

```python
names["artist"].unique() #checking if new df contains names of only top 10 occuring artisst
```

```python
df_names = names.groupby("artist") #using groupby group all same names together to perform mathematical ops on it
```

```python
df_names = df_names.mean() #calculating average of all numerical values of top 10 artists
```

```python
df_names_final = df_names[["tempo"]] #making a df containing average of only tempo of top 10 artists
```
